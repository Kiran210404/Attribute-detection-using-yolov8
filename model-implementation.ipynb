```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
from ultralytics import YOLO

# 1) Load YOLO Model (Custom-trained weights)
yolo_model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

# 2) Define your custom classes (in English)
class_names = {
    0: "Obstructed Path",
    1: "Inaccessible Due to Construction",
    2: "Ramp Needed",
    3: "Inaccessible Ramp",
    4: "Unstable Ground"
}

# 3) Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(
    static_image_mode=False,
    model_complexity=2,
    enable_segmentation=False
)
mp_drawing = mp.solutions.drawing_utils

def detect_objects_and_pose(image_path):
    # Read image in BGR
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found at {image_path}")

    # Convert to RGB for MediaPipe
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # --- YOLO Inference ---
    results = yolo_model(image_rgb)
    # Extract bounding boxes
    for box in results[0].boxes.data.cpu().numpy():
        x1, y1, x2, y2, conf, cls = box
        
        # Use your custom class names
        class_name = class_names.get(int(cls), "Unknown")
        
        # Draw YOLO bounding boxes on BGR image
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
        cv2.putText(
            image,
            f"{class_name} {conf:.2f}",
            (int(x1), int(y1) - 5),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0,255,0),
            2
        )

    # --- MediaPipe Pose ---
    pose_results = pose.process(image_rgb)
    if pose_results.pose_landmarks:
        mp_drawing.draw_landmarks(
            image,  # Draw on BGR
            pose_results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS
        )

    # Show final integrated image
    plt.figure(figsize=(8,6))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title("YOLO + MediaPipe Pose with Custom Classes")
    plt.show()

# 4) Test on a sample image
test_image = "/kaggle/input/dataset1/train/images/000001683_jpg.rf.15ebd6cece187b2c8710186adebc1f4c.jpg"  # Adjust path
detect_objects_and_pose(test_image)
```

```python
def compute_scale_factor(person_box, real_height=1.7):
    """
    person_box: [x1, y1, x2, y2, conf, cls] for a 'person'
    real_height: approximate real-world height in meters
    returns: scale_factor (meters per pixel)
    """
    x1, y1, x2, y2, _, _ = person_box
    pixel_height = (y2 - y1)
    if pixel_height <= 0:
        return None
    scale_factor = real_height / pixel_height
    return scale_factor
```

```python
def pixel_to_meters(pixel_distance, scale_factor):
    return pixel_distance * scale_factor
```

```python
def compute_ramp_slope(ramp_box, scale_factor):
    """
    ramp_box: [x1, y1, x2, y2, conf, cls]
    scale_factor: meters per pixel
    returns: slope (height/length)
    """
    x1, y1, x2, y2, _, _ = ramp_box
    pixel_height = (y2 - y1)  # approximate ramp 'vertical'
    pixel_length = (x2 - x1)  # approximate ramp 'horizontal'
    
    ramp_height_m = pixel_height * scale_factor
    ramp_length_m = pixel_length * scale_factor
    
    if ramp_length_m == 0:
        return float('inf')
    slope = ramp_height_m / ramp_length_m
    return slope
```

```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
from ultralytics import YOLO

# YOLO Model (custom-trained)
yolo_model = YOLO("/kaggle/working/runs/detect/train/weights/best.pt")

# Class names
class_names = {
    0: "Obstructed Path",
    1: "Inaccessible Due to Construction",
    2: "Ramp Needed",
    3: "Inaccessible Ramp",
    4: "Unstable Ground"
}

# MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(
    static_image_mode=False,
    model_complexity=2,
    enable_segmentation=False
)
mp_drawing = mp.solutions.drawing_utils

# Hardcoded scale factor (example)
SCALE_FACTOR = 0.005  # 0.005 m per pixel

def interpret_ramp_slope(slope):
    """
    Returns a suggestion based on the slope:
      < 0.05  -> Very gentle
      0.05-0.083 -> Borderline
      > 0.083 -> Too steep (1:12 slope standard)
    """
    if slope < 0.05:
        return "Ramp is fairly gentle, good accessibility!"
    elif slope < 0.083:  # borderline
        return "Ramp is borderline, caution recommended."
    else:
        return "Ramp is too steep! Consider alternative route or fix."

def compute_ramp_slope(ramp_box, scale_factor):
    x1, y1, x2, y2, _, _ = ramp_box
    pixel_height = (y2 - y1)  # approximate vertical dimension
    pixel_length = (x2 - x1)  # approximate horizontal dimension
    
    ramp_height_m = pixel_height * scale_factor
    ramp_length_m = pixel_length * scale_factor
    
    if ramp_length_m == 0:
        return float('inf')
    return ramp_height_m / ramp_length_m

def detect_objects_and_pose(image_path):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image not found at {image_path}")

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # --- YOLO Inference ---
    results = yolo_model(image_rgb)
    boxes = results[0].boxes.data.cpu().numpy()
    
    # We'll collect ramp boxes for slope calculation
    ramp_boxes = []

    for box in boxes:
        x1, y1, x2, y2, conf, cls = box
        cls_id = int(cls)
        class_name = class_names.get(cls_id, "Unknown")
        
        # Draw bounding boxes
        cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
        cv2.putText(image, f"{class_name} {conf:.2f}", (int(x1), int(y1) - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)
        
        # If ramp is detected
        if class_name in ["Ramp Needed", "Inaccessible Ramp"]:
            ramp_boxes.append(box)
    
    # --- MediaPipe Pose ---
    pose_results = pose.process(image_rgb)
    if pose_results.pose_landmarks:
        mp_drawing.draw_landmarks(
            image,
            pose_results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS
        )
    
    # --- Compute slope for each ramp using the hardcoded scale ---
    for ramp_box in ramp_boxes:
        slope = compute_ramp_slope(ramp_box, SCALE_FACTOR)
        slope_message = interpret_ramp_slope(slope)
        print(f"Ramp slope: {slope:.2f}. {slope_message}")
    
    # Show final integrated image
    plt.figure(figsize=(8,6))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.title("YOLO + MediaPipe Pose + Hardcoded Scale Factor")
    plt.show()

# Test the function
test_image = "/kaggle/input/my-example/train/34a2c644-images_11_jpeg.rf.4bdb3909131e711cfea459911e7b53b8.jpg"
detect_objects_and_pose(test_image)
```
